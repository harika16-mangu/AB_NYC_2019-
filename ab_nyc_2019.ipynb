```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data =pd.read_csv("C:/Users/harik/Downloads/scholarship.csv")
data.head()
```

```python
data1=data.copy()
data1.head()
```

data1.shape

```python
plt.subplot(1,2,1)
sns.distplot(data1["semester_percentage"])
plt.subplot(1,2,2)
sns.distplot(data1["scholarship_exam_marks"])
plt.show()
```

#first plot is the normal distribution and second plot is left skewed distribution

```python
mean=data1["semester_percentage"].mean()
std=data1["semester_percentage"].std()
```

```python
print(mean,std)
```

```python
highest_allowed=mean+3*std
lowest_allowed=mean-3*std
```

```python
print(highest_allowed," ",lowest_allowed)
```

```python
data1[(data1["semester_percentage"]<lowest_allowed) | (data1["semester_percentage"]>highest_allowed)]
```

Trimming

```python
#data without outliers
data1[(data1["semester_percentage"]>lowest_allowed) & (data1["semester_percentage"]<highest_allowed)]
```

 Trimming using z_scores

```python
#z_score=(x-mean)/std
data1["z_score"]=(data1["semester_percentage"]-data1["semester_percentage"].mean())/data1["semester_percentage"].std()
```

```python
data1.head()
```

```python
data1[(data1["z_score"]<-3) | (data1["z_score"]>3)]
```

```python
data1[(data1["z_score"]>-3) & (data1["z_score"]<3)]
```

Capping 

```python
import numpy as np
upper_limit=mean+3*std
lower_limit=mean-3*std
print(upper_limit," ",lower_limit)
```

```python
data1["semester_percentage"]=np.where(data1["semester_percentage"]>upper_limit,upper_limit,data1["semester_percentage"])
data1["semester_percentage"]=np.where(data1["semester_percentage"]<lower_limit,lower_limit,data1["semester_percentage"])
```

```python
data1.head()
```

```python
data1.shape
```

```python
data1[(data1["semester_percentage"]<lower_limit) | (data1["semester_percentage"]>upper_limit)]
```

```python
data1
```

```python
#scholarship_exam_marks-interquartile range
```

The data is divided into 4 equal parts which are known as quartiles and each quartile has 25% of data.
1.min_value to Q1- Interval 1
2.Q1-Q2
3.Q2-Q3
4.Q3-max_value
Interquartile range concept is used to solve used when the distribution of the data is skewed
Q3-Q1 is interquartile
values less than q1-1.5*iqr and q3+1.5*iqr are considered to be outliers

```python
q1=data1["scholarship_exam_marks"].quantile(0.25)
q2=data1["scholarship_exam_marks"].quantile(0.50)
q3=data1["scholarship_exam_marks"].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
iqr=q3-q1
print(iqr)
```

```python
data1.describe()
```

```python
min=q1-1.5*iqr
max=q3+1.5*iqr
print(min," ",max)
```

```python
data1[(data1["scholarship_exam_marks"]>max) |(data1["scholarship_exam_marks"]<min)]
```

```python
#Trimming
```

```python
new_data=data1[(data1["scholarship_exam_marks"]>min) & (data1["scholarship_exam_marks"]<max)]
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(data1["scholarship_exam_marks"])

plt.subplot(2,2,2)
sns.boxplot(data1["scholarship_exam_marks"])

plt.subplot(2,2,3)
sns.distplot(new_data["scholarship_exam_marks"])

plt.subplot(2,2,4)
sns.boxplot(new_data["scholarship_exam_marks"])

plt.show()
```

```python
data1["scholarship_exam_marks"]=np.where(data1["scholarship_exam_marks"]>max ,max,np.where(data1["scholarship_exam_marks"]<min,min,data1["scholarship_exam_marks"]))
```
